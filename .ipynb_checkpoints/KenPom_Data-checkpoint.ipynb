```python
import pandas as pd
import numpy as np
```

```python
for year in range(2002,2019):
    pt_filename = "KenPom"+str(year)+"_pt.csv"
    final_filename = "KenPom_Final/KenPom"+str(year)+".csv"
    pt_kenpom = pd.read_csv(pt_filename,index_col=0)
    final_kenpom = pd.read_csv(final_filename,index_col=0)
    
    pt_kenpom = pt_kenpom.sort_values(by=['TeamName'])
    final_kenpom = final_kenpom.sort_values(by=['Team'])

    pt_kenpom = pt_kenpom.reset_index()
    final_kenpom = final_kenpom.reset_index()
    
    for n in range(len(pt_kenpom['TeamName'])):
        team_a = pt_kenpom['TeamName'][n]
        team_b = final_kenpom['Team'][n][:-1]
        if team_a != team_b:
            print(year, "team names don't line up,",team_a,team_b)
            continue
    
    pt_kenpom['SOSAdjEM'] = final_kenpom['SOSAdjEM']
    pt_kenpom['NCSOSAsjEM'] = final_kenpom['NCSOSAdjEM']
    
    pt_kenpom = pt_kenpom.set_index('index')
    pt_kenpom = pt_kenpom.sort_values(by='AdjEM',ascending=False)
    
    pt_kenpom.to_csv(pt_filename)
    print(year, "written")
```

```python
for year in range(2002,2019):
    ofilename = "FourFactorData/4f_offense"+str(year)+".csv"
    dfilename = "FourFactorData/4f_defense"+str(year)+".csv"
    kpfilename = "KenPom"+str(year)+"_pt.csv"

    ff_o = pd.read_csv(ofilename)
    ff_d = pd.read_csv(dfilename)
    kenpom = pd.read_csv(kpfilename)

    teamlist = kenpom['TeamName']
    teamlist = list(teamlist)

    for i in range(len(ff_o['TeamName'])):
        if ff_o['TeamName'][i] in teamlist:
            pass
        else:
            ff_o = ff_o.drop([i])

    for i in range(len(ff_d['TeamName'])):
        if ff_d['TeamName'][i] in teamlist:
            pass
        else:
            ff_d = ff_d.drop([i])
    
    kenpom = kenpom.sort_values(by=['TeamName'])
    ff_o = ff_o.sort_values(by=['TeamName'])
    ff_d = ff_d.sort_values(by=['TeamName'])

    kenpom = kenpom.reset_index()
    kenpom = kenpom.drop(columns=['level_0'])
    ff_o = ff_o.reset_index()
    ff_d = ff_d.reset_index()
        
    kenpom['O_eFG_Pct'] = ff_o['eFG_Pct']
    kenpom['D_eFG_Pct'] = ff_d['eFG_Pct']
    kenpom['O-D_eFG_Pct'] = kenpom['O_eFG_Pct'] - kenpom['D_eFG_Pct']

    kenpom['O_TO_Pct'] = ff_o['TO_Pct']
    kenpom['D_TO_Pct'] = ff_d['TO_Pct']
    kenpom['D-O_TO_Pct'] = kenpom['D_TO_Pct'] - kenpom['O_TO_Pct']

    kenpom['O_OR_Pct'] = ff_o['OR_Pct']
    kenpom['D_OR_Pct'] = ff_d['OR_Pct']
    kenpom['O-D_OR_Pct'] = kenpom['O_OR_Pct'] - kenpom['D_OR_Pct']

    kenpom['O_FT_Rate'] = ff_o['FT_Rate']
    kenpom['D_FT_Rate'] = ff_d['FT_Rate']
    kenpom['O-D_FT_Rate'] = kenpom['O_FT_Rate'] - kenpom['D_FT_Rate']
    
    kenpom = kenpom.set_index('index')
    kenpom = kenpom.sort_values(by='AdjEM',ascending=False)
    
    kenpom.to_csv(kpfilename)
    print(year, "written")
```

```python
# remove punctuation from KenPom Data
import string
translator = str.maketrans('', '', string.punctuation)

names_df = pd.read_csv("TeamNameInconsistencies.csv")
kenpom_names = list(names_df['KenPom'])
results_names = list(names_df['Results'])

for year in range(2002,2019):
    filename = "KenPom"+str(year)+"_pt.csv"
    kp = pd.read_csv(filename)
    for i in range(len(kp['TeamName'])):
        kp['TeamName'][i] = kp['TeamName'][i].translate(translator)
        
        if kp['TeamName'][i] in kenpom_names: #Change KenPom names to Results names
            indx = kenpom_names.index(kp['TeamName'][i])
            kp['TeamName'][i] = results_names[indx]
    kp.to_csv(filename)
    print(year, "written")
```

```python
results = pd.read_csv("full_results.csv")

for year in range(2002,2019):
    kpfilename = "KenPom"+str(year)+"_pt.csv"
    kp = pd.read_csv(kpfilename)
    
    teamlist = []
    ctr = 0
    
    for i in range(len(results['TeamA'])):
        if results['Year'][i] == year:
            if results['TeamA'][i] not in teamlist:
                teamlist.append(results['TeamA'][i])
            if results['TeamB'][i] not in teamlist:
                teamlist.append(results['TeamB'][i])
    kpteamlist = sorted(list(kp['TeamName']))
    teamlist = sorted(teamlist)
    print("\nYear:",year,"  ",len(kpteamlist),len(teamlist))
    for k in range(len(teamlist)):
        if kpteamlist[k] != teamlist[k]:
            print(kpteamlist[k],"  ",teamlist[k])
            ctr += 1
    if ctr == 0:
        print("All match")
```

```python
full_results = pd.read_csv("Kaggle/RegularSeasonCompactResults.csv")
teamnames = pd.read_csv("Kaggle/Teams.csv")
kp_full = pd.read_csv("FullKenPom_pt.csv")
```

```python
kp_full['LastTenRecord'] = pd.Series(np.zeros(len(kp_full['TeamName'])), index=kp_full.index)
kp_full.head()
```

```python
for year in range(2002,2019): # do this by year
    filename = "KenPom"+str(year)+"_pt.csv"
    kp_yr = pd.read_csv(filename)
    
    record_dict = {}
    k=0
    
    for team in kp_yr['TeamName']:
        k += 1
        wins = 0
        losses = 0
        for i in range(len(teamnames['TeamName'])):
                if teamnames['TeamName'][i] == team:
                    team_id = teamnames['TeamID'][i]
        for i in range(len(full_results['WTeamID'])-1,0,-1):
            if (wins+losses) == 10:
                break
            if full_results['Season'][i] == year:
                if full_results['WTeamID'][i] == team_id:
                    wins += 1
                elif full_results['LTeamID'][i] == team_id:
                    losses += 1
        record = wins/10
        for i in range(len(kp_full['TeamName'])):
            if (kp_full['Season'][i] == year) and (kp_full['TeamName'][i] == team):
                kp_full['LastTenRecord'][i] = record
        
        record_dict[team] = [wins,losses,record]
        print(year,"  ",k)
        
    print("Year:",year)
    print(record_dict)
    print("\n\n")
```
